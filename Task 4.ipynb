## Data Visualizing

### Countries affected by terrorist attacks

```python
df['country'].value_counts()
plt.figure(figsize=(16,10))
sns.barplot(x=df['country_txt'].value_counts()[:30].index ,y=df['country_txt'].value_counts()[:30].values)
plt.title('TOP COUNTRIES GOT ATTACKS', fontsize=15)
plt.xticks(rotation=90)
plt.xlabel('Countries', fontsize=15)
plt.ylabel('Number of attacks', fontsize=15)
plt.show()
```

### Region affected by terrorist attacks

```python
plt.figure(figsize=(16,10))
sns.barplot(x=df['region_txt'].value_counts().index ,y=df['region_txt'].value_counts().values)
plt.title('REGIONS AFFECTED BY ATTACKS', fontsize=15)
plt.xticks(rotation=70)
plt.xlabel('Regions', fontsize=15)
plt.ylabel('Number of attacks', fontsize=15)
plt.show()
```

### Types of attack used by terrorist

```python
plt.figure(figsize=(16,10))
sns.barplot(x=df['attacktype1_txt'].value_counts().index ,y=df['attacktype1_txt'].value_counts().values)
plt.title('Types of attack used by terrorist', fontsize=15)
plt.xticks(rotation=70)
plt.xlabel('Attack type', fontsize=15)
plt.ylabel('Number of attacks', fontsize=15)
plt.show()
```

### Years with most terrorist activity

```python
plt.figure(figsize=(16,10))
sns.barplot(x=df['iyear'].value_counts().index ,y=df['iyear'].value_counts().values, palette='rainbow')
plt.title('YEARS WITH MOST TERRORIST ACTIVITY', fontsize=15)
plt.xticks(rotation=90)
plt.xlabel('Year', fontsize=15)
plt.ylabel('Number of Attacks', fontsize=15)
plt.show()
```

### Attack target

```python
plt.figure(figsize=(16,10))
sns.barplot(x=df['targtype1_txt'].value_counts().index ,y=df['targtype1_txt'].value_counts().values, palette='rainbow')
plt.title('TARGETS', fontsize=15)
plt.xticks(rotation=90)
plt.xlabel('Targets', fontsize=15)
plt.ylabel('Number of Targets', fontsize=15)
plt.show()
```

### Name of group attacks terrorist

```python
plt.figure(figsize=(16,10))
sns.barplot(x=df['gname'].value_counts()[:20].index ,y=df['gname'].value_counts()[:20].values)
plt.title('Total group attacks', fontsize=15)
plt.xticks(rotation=90)
plt.xlabel('Group name', fontsize=15)
plt.ylabel('Total Number of attacks', fontsize=15)
plt.show()
```

### Cities affected by terrorist

```python
plt.figure(figsize=(16,10))
sns.barplot(x=df['city'].value_counts()[:30].index ,y=df['city'].value_counts()[:30].values, palette='rainbow')
plt.title('TOP CITIES GOT ATTACKS', fontsize=15)
plt.xticks(rotation=90)
plt.xlabel('Cities', fontsize=15)
plt.ylabel('Number of attacks', fontsize=15)
plt.show()
```

### Weapon used for supporting the attack type

```python
plt.figure(figsize=(16,10))
sns.barplot(x=df['weaptype1_txt'].value_counts()[:30].index ,y=df['weaptype1_txt'].value_counts()[:30].values)
plt.title('WEAPON USAGS', fontsize=15)
plt.xticks(rotation=90)
plt.xlabel('Weapon used', fontsize=15)
plt.ylabel('Number of times', fontsize=15)
plt.show()
```

### Kill - wound visualization

```python
df.plot(kind = 'scatter', x = 'nkill', y = 'nwound', alpha = 0.5, color = 'r', figsize = (16,10), fontsize=15)
plt.xlabel('Kill', fontsize=15)
plt.ylabel('Wound', fontsize=15)
plt.title('Kill - Wound Scatter Plot',fontsize=15)
plt.show()
```
